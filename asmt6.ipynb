```python
import pandas as pd
import numpy as np

df = pd.DataFrame({'From_To': ['LoNDon_paris', 'MAdrid_miLAN',
'londON_StockhOlm',
'Budapest_PaRis', 'Brussels_londOn'],
'FlightNumber': [10045, np.nan, 10065, np.nan, 10085],
'RecentDelays': [[23, 47], [], [24, 43, 87], [13], [67, 32]],
'Airline': ['KLM(!)', '<Air France> (12)', '(British Airways. )',
'12. Air France', '"Swiss Air"']})
```

```python
df
```

```python
for i in range(len(df)):
    if np.isnan(df['FlightNumber'][i]):
        df.loc[i,('FlightNumber')] = df.loc[i-1, ('FlightNumber')] +10
df['FlightNumber'] = df['FlightNumber'].astype(int)
        
```

```python
df
```

```python
df_tmp = pd.DataFrame()
df_tmp[['From','To']] = df.From_To.str.split('_', expand = True)
```

```python
for column in df_tmp:
    df_tmp[column] = df_tmp[column].str.capitalize()
```

```python
df_tmp
```

```python
df = df.drop('From_To', 1)
```

```python
df = df.join(df_tmp)
```

```python
df
```

```python
delays = pd.DataFrame(df['RecentDelays'].to_list(), columns=['delay_1', 'delay_2', 'delay_3'])
```
